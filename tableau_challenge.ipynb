```python
import pandas as pd
import datetime as dt

```

```python
# Read CSV files
df1 = pd.read_csv('./202312-citibike-tripdata_3.csv')
df2 = pd.read_csv('./JC-202212-citibike-tripdata.csv')
```

```python
# To convert the starter at to datetime so that the year can be extracted
df1['started_at'] = pd.to_datetime(df1["started_at"])
df1.head()
```

```python
# To convert the starter at to datetime so that the year can be extracted
df2["started_at"] = pd.to_datetime(df2["started_at"])
df2.head()
```

```python
# To extract year of dataframe  # 1
df1["year"] = df1["started_at"].dt.year
df1.head()
```

```python
# To extract year of dataframe  # 2
df2["year"] = df2["started_at"].dt.year
df2.head()
```

```python
# To combine both dataframes
merged_df = pd.concat([df1, df2], axis = 0)
```

```python
merged_df.head()
```

## Question 1

```python
# Extract total number of unique bikes by year by type
type_bikes = merged_df.groupby(['year','rideable_type']).count()
type_bikes_reset = type_bikes.reset_index()
type_bikes_reset
```

```python
# to drop unwanted columns
type_bikes_cleaned = type_bikes_reset.drop(["started_at", "ended_at", "start_station_name", "start_station_id", "end_station_name", "end_station_id", "start_lat", "start_lng", "end_lat", "end_lng", "member_casual"], axis = 1)
type_bikes_cleaned.rename (columns={"ride_id": "total_bikes"}, inplace= True)
type_bikes_cleaned
```

```python
# to export the final csv file
type_bikes_cleaned.to_csv("output_files/type_bikes_cleaned.csv", index =False)
```

## Question 2

```python
unique_values = merged_df["start_station_id"].unique()
unique_values
```

```python
# Extract total number of unique stations by year by total bikes
start_stations = merged_df.groupby(['year','start_station_id']).count()
start_stations = start_stations.reset_index()
start_stations
```

```python
# to drop unwanted columns
start_stations_cleaned = start_stations.drop(["rideable_type", "ended_at", "start_station_name", "end_station_name", "end_station_id", "start_lat", "start_lng", "end_lat", "end_lng", "member_casual"], axis = 1)
start_stations_cleaned.rename (columns={"ride_id": "total_bikes"}, inplace= True)
start_stations_cleaned
```

```python
# to export the final csv file
start_stations_cleaned.to_csv("output_files/start_stations_cleaned.csv", index =False)
```
